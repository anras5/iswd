```python
CRITERIA = {
    'PTS': {'type': 'gain', 'k': 6,    'q': 3,   'p': 5,   'v': 8},
    'FG%': {'type': 'gain', 'k': 3.14, 'q': 3,   'p': 4,   'v': 8},
    '3P%': {'type': 'gain', 'k': 3.14, 'q': 4,   'p': 5,   'v': 11},
    'FT%': {'type': 'gain', 'k': 1.71, 'q': 4,   'p': 8,   'v': 13},
    'REB': {'type': 'gain', 'k': 3.86, 'q': 2,   'p': 5,   'v': 9},
    'AST': {'type': 'gain', 'k': 3.86, 'q': 2,   'p': 4,   'v': 7},
    'TOV': {'type': 'cost', 'k': 2.43, 'q': 1,   'p': 1.5, 'v': 2},
    'PF':  {'type': 'cost', 'k': 1,    'q': 1,   'p': 1.5, 'v': 2},
    '+/-': {'type': 'gain', 'k': 1.71, 'q': 3,   'p': 4,   'v': 8},
}
```

## Marginal concordance and discordance

```python
def calculate_marginal_concordance(ga, gb, q, p, type_='gain'):

    if type_ == 'gain':
        if ga - gb >= -q:
            return 1
        elif ga - gb < -p:
            return 0
        else:
            return (p-(gb-ga))/(p-q)
    elif type_ == 'cost':
        if ga - gb <= q:
            return 1
        elif ga - gb > p:
            return 0
        else:
            return (p-(ga-gb))/(p-q)
    else:
        raise Exception("wrong criterion type")
    

def calculate_marginal_discordance(ga, gb, p, v, type_='gain'):

    if type_ == 'gain':
        if ga - gb <= -v:
            return 1
        elif ga - gb >= -p:
            return 0
        else:
            return ((gb-ga)-p)/(v-p)
    elif type_ == 'cost':
        if ga - gb >= v:
            return 1
        elif ga - gb <= p:
            return 0
        else:
            return (v-(ga-gb))/(v-p)
    else:
        raise Exception("wrong criterion type")
```

## Comprehensive Concordance

```python
# Comprehensive Concordance will be stored inside df_C variable which is a NxN dataframe
df_C = pd.DataFrame(1.0, index=df.index, columns=df.index)
weights = sum([config['k'] for config in CRITERIA.values()])

for alt_1_id in df_C.index:
    for alt_2_id in df_C.index:
        if alt_1_id == alt_2_id:
            continue

        comprehensive_concordance_index = 0
        # calculate marginal concordance index for each criterion and add it to the cc index with appropriate weight
        for criterion, config in CRITERIA.items():
            marginal_concordance = calculate_marginal_concordance(
                ga=df.loc[alt_1_id, criterion],
                gb=df.loc[alt_2_id, criterion],
                q=config['q'],
                p=config['p'],
                type_=config['type']
            )
            comprehensive_concordance_index += config['k'] * marginal_concordance
        # normalize the cc index
        comprehensive_concordance_index /= weights
        df_C.loc[alt_1_id, alt_2_id] = comprehensive_concordance_index
```

## Outranking Credibility

```python
df_sigma = pd.DataFrame(1.0, index=df.index, columns=df.index)

for alt_1_id in df_C.index:
    for alt_2_id in df_C.index:
        if alt_1_id == alt_2_id:
            continue

        C = sigma = df_C.loc[alt_1_id, alt_2_id]
        # calculate marginal discordance for each criterion
        for criterion, config in CRITERIA.items():
            marginal_discordance = calculate_marginal_discordance(
                ga=df.loc[alt_1_id, criterion],
                gb=df.loc[alt_2_id, criterion],
                p=config['p'],
                v=config['v'],
                type_=config['type']
            )
            # if the marginal discordance is greater than the result of concordance test,
            # take it into account while calculating outranking credibility (sigma)
            if marginal_discordance > C:
                sigma = sigma * (1-marginal_discordance)/(1-C)
        df_sigma.loc[alt_1_id, alt_2_id] = sigma

np.fill_diagonal(df_sigma.values, -1)
```

## Distillation

```python
def distillation(df_m, upper_threshold=None, descending=True):

    # compute upper threshold
    if upper_threshold is None:
        upper_threshold = df_m.values.max()
    
    # compute lower threshold
    s = -0.15 * upper_threshold + 0.3
    lower_threshold = upper_threshold - s
    lower_threshold = df_m[(df_m < lower_threshold) & (df_m > -1)].max().max()
    if np.isnan(lower_threshold):
        lower_threshold = 0
    
    # print(df_m)
    # print(f"{s=}, {lower_threshold=}, {type(lower_threshold)} {upper_threshold=}")

    # if upper threshold is equal to 0, then stop the algorithm
    if upper_threshold == 0:
        return df_m.index

    # save only relations that satisfy given conditions to the df_f (df filtered)
    df_f = pd.DataFrame(-1.0, index=df_m.index, columns=df_m.index)
    for alt_1_id in df_m.index:
        for alt_2_id in df_m.columns:
            if alt_1_id == alt_2_id:
                continue
            sigma = df_m.loc[alt_1_id, alt_2_id]
            if sigma > lower_threshold and sigma > df_m.loc[alt_2_id, alt_1_id] + s:
                df_f.loc[alt_1_id, alt_2_id] = sigma

    # compute the strength, weakness and quality
    df_swq = pd.DataFrame(0, index=df_m.index, columns=['strength', 'weakness', 'quality'])
    df_swq['strength'] = df_f.apply(lambda row: (row != -1.0).sum(), axis=1)
    df_swq['weakness'] = df_f.apply(lambda col: (col != -1.0).sum(), axis=0)
    df_swq['quality'] = df_swq['strength'] - df_swq['weakness']

    # take subset with maximal quality
    df_swq['rank'] = df_swq['quality'].rank(ascending=False, method='min')
    if descending:
        df_mq = df_swq[df_swq['rank'] == 1]
    else:
        df_mq = df_swq[df_swq['rank'] == df_swq['rank'].max()]
    if len(df_mq.index) == 1:
        return df_mq.index
    else:
        return distillation(
            df_m.loc[df_mq.index, df_mq.index],
            upper_threshold=lower_threshold,
            descending=descending
        )
```

## Adding descending preorder and ascending preorder to the main dataframe

```python
df['desc_preorder'] = 0
df['asc_preorder'] = 0

descending_results = []
df_copy = df_sigma.copy()
position = 1
while len(df_copy.index) > 0:
    alternatives = distillation(df_copy, descending=True)
    df_copy = df_copy.loc[~df_copy.index.isin(alternatives), ~df_copy.index.isin(alternatives)]
    df.loc[alternatives, 'desc_preorder'] = position
    descending_results.append([df.loc[alt, "PLAYER"] for alt in alternatives])
    position += 1
print(f"{descending_results=}")

ascending_results = []
df_copy = df_sigma.copy()
position = len(df_sigma)
while len(df_copy.index) > 0:
    alternatives = distillation(df_copy, descending=False)
    df_copy = df_copy.loc[~df_copy.index.isin(alternatives), ~df_copy.index.isin(alternatives)]
    df.loc[alternatives, 'asc_preorder'] = position
    ascending_results.insert(0, [df.loc[alt, "PLAYER"] for alt in alternatives])
    position -= 1
df['asc_preorder'] = df['asc_preorder'] - df['asc_preorder'].min() + 1
print(f"{ascending_results=}")
```

```python
df_final = pd.DataFrame(0, index=df.index, columns=df.index)

for alt_1_id in df_final.index:
    for alt_2_id in df_final.columns:
        if alt_1_id == alt_2_id:
            continue

        alt_1_d = df.loc[alt_1_id, "desc_preorder"]
        alt_1_a = df.loc[alt_1_id, "asc_preorder"]
        alt_2_d = df.loc[alt_2_id, "desc_preorder"]
        alt_2_a = df.loc[alt_2_id, "asc_preorder"]

        if alt_1_d < alt_2_d and alt_1_a < alt_2_a:
            df_final.loc[alt_1_id, alt_2_id] = 1
        elif (alt_1_d < alt_2_d and alt_1_a == alt_2_a) or (alt_1_d == alt_2_d and alt_1_a < alt_2_a):
            df_final.loc[alt_1_id, alt_2_id] = 1
        elif alt_1_d == alt_2_d and alt_1_a == alt_2_a:
            df_final.loc[alt_1_id, alt_2_id] = 2

# ---------------------------------------------------------------------------------------------------- #
# Code used to produce output for graph visualization
            
# relations = []
# for alt_1_id in df_final.index:
#     for alt_2_id in df_final.columns:
#         if alt_1_id == alt_2_id:
#             continue
#         if df_final.loc[alt_1_id, alt_2_id]:
#             relations.append({"alternative_1": alt_1_id, "alternative_2": alt_2_id})

# alternatives = []
# for alt_1_id in df.index:
#     alternatives.append({"id": alt_1_id, "name": df.loc[alt_1_id, "PLAYER"]})

# print(relations)
# print(alternatives)
```

## Finding rank for each alternative

```python
def find_longest_paths(df):
    def dfs(node, path):
        if node not in parents:
            return
        for parent in parents[node]:
            dfs(parent, path + [parent])
        longest_paths[node] = max(longest_paths[node], path, key=len)

    # find parents for each node
    parents = {}
    for row in df.index:
        parents[row] = [col for col in df.columns if df.loc[row, col] == 1]

    # initialize path for each node with empty list
    longest_paths = {node: [] for node in df.columns}

    # perform DFS from each node
    for node in df.columns:
        dfs(node, [])

    return longest_paths

longest_paths = find_longest_paths(df_final)
df['ranks'] = [len(path)+1 for path in longest_paths.values()]
```

## Finding median rank for each alternative

```python
df['median'] = 0
position = 1
for rank_value in df['ranks'].sort_values().unique():

    rank_group = df[df['ranks'] == rank_value].copy()
    rank_group['plus'] = rank_group['desc_preorder'] + rank_group['asc_preorder']
    rank_group = rank_group.sort_values('plus')
    for unique_plus in rank_group['plus'].unique():
        rank_group_unique_plus = rank_group[rank_group['plus'] == unique_plus]
        for alt_id in rank_group_unique_plus.index:
            df.loc[alt_id, "median"] = position

        position += 1
```

## Comprehensive Preference Index

```python
# Comprehensive Preference Index will be stored inside df_C variable which is a NxN dataframe
df_PI = pd.DataFrame(1.0, index=df.index, columns=df.index)
weights = sum([config['k'] for config in PROMETHEE.values()])

for alt_1_id in df_PI.index:
    for alt_2_id in df_PI.index:
        if alt_1_id == alt_2_id:
            continue

        comprehensive_preference_index = 0
        # calculate marginal preference index for each criterion and add it to the cp index with appropriate weight
        for criterion, config in PROMETHEE.items():
            marginal_preference = calculate_marginal_preference(
                ga=df.loc[alt_1_id, criterion],
                gb=df.loc[alt_2_id, criterion],
                q=config['q'],
                p=config['p'],
                type_=config['type']
            )
            comprehensive_preference_index += config['k'] * marginal_preference
        # normalize the cp index
        comprehensive_preference_index /= weights
        df_PI.loc[alt_1_id, alt_2_id] = comprehensive_preference_index
```

## Calculating Positive and Negative Flows and PROMETHEE rankings

```python
# calculating flows
df['P_negative_flow'] = df_PI.sum()
df['P_positive_flow'] = df_PI.sum(axis=1)
df['P_balance'] = df['P_positive_flow'] - df['P_negative_flow']

# calculating negative and positive ranking
df['P_ranking_negative'] = df['P_negative_flow'].rank(method='min').astype(int)
df['P_ranking_positive'] = df['P_positive_flow'].rank(ascending=False, method='min').astype(int)

df_ranking_promethee_I = pd.DataFrame(0, index=df.index, columns=df.index)
for alt_1_id in df_ranking_promethee_I.index:
    for alt_2_id in df_ranking_promethee_I.columns:
        if alt_1_id == alt_2_id:
            continue

        alt_1_n = df.loc[alt_1_id, 'P_negative_flow']
        alt_1_p = df.loc[alt_1_id, 'P_positive_flow']
        alt_2_n = df.loc[alt_2_id, 'P_negative_flow']
        alt_2_p = df.loc[alt_2_id, 'P_positive_flow']
        # preference
        if alt_1_p >= alt_2_p and alt_1_n <= alt_2_n:
            df_ranking_promethee_I.loc[alt_1_id, alt_2_id] = 1

df_ranking_promethee_II = pd.DataFrame(0, index=df.index, columns=df.index)
for alt_1_id in df_ranking_promethee_II.index:
    for alt_2_id in df_ranking_promethee_II.columns:
        if alt_1_id == alt_2_id:
            continue

        if df.loc[alt_1_id, 'P_balance'] >= df.loc[alt_2_id, 'P_balance']:
            df_ranking_promethee_II.loc[alt_1_id, alt_2_id] = 1

# ---------------------------------------------------------------------------------------------------- #
# Code used to produce output for graph visualization

# PROMETHEE I
# print(alternatives)

# relations = []
# for alt_1_id in df_ranking_promethee_I.index:
#     for alt_2_id in df_ranking_promethee_I.columns:
#         if alt_1_id == alt_2_id:
#             continue
#         if df_ranking_promethee_I.loc[alt_1_id, alt_2_id]:
#             relations.append({"alternative_1": alt_1_id, "alternative_2": alt_2_id})

# alternatives = []
# for alt_1_id in df.index:
#     alternatives.append({"id": alt_1_id, "name": df.loc[alt_1_id, "PLAYER"]})

# print(relations)

# # PROMETHEE II
# relations = []
# for alt_1_id in df_ranking_promethee_II.index:
#     for alt_2_id in df_ranking_promethee_II.columns:
#         if alt_1_id == alt_2_id:
#             continue
#         if df_ranking_promethee_II.loc[alt_1_id, alt_2_id]:
#             relations.append({"alternative_1": alt_1_id, "alternative_2": alt_2_id})

# print(relations)
```
