8. Jakie są wagi poszczególnych kryteriów?

Aby ustalić wagi danych kryteriów zastosowana została procedura SRF. Poniżej zaprezentowane zostało ustawienie kart:

`Ratio Z` ustawione zostało na 6. Po wykonaniu obliczeń wagi przedstawiają się nastepująco:

| CRITERION | WEIGHT |
| --------- | ------ |
| PTS | 6 |
| AST, REB | 3.86 |
| FG%, 3P% | 3.14 |
| TOV | 2.43 |
| FT%, +/- | 1.71 |
| PF | 1 |

```python
CRITERIA = {
    'PTS': {'type': 'gain', 'k': 6, 'q': 4, 'p': 8, 'v': 12},
    'FG%': {'type': 'gain', 'k': 3.14, 'q': 3, 'p': 4, 'v': 8},
    '3P%': {'type': 'gain', 'k': 3.14, 'q': 4, 'p': 5, 'v': 11},
    'FT%': {'type': 'gain', 'k': 1.71, 'q': 3, 'p': 7, 'v': 15},
    'REB': {'type': 'gain', 'k': 3.86, 'q': 2, 'p': 5, 'v': 9},
    'AST': {'type': 'gain', 'k': 3.86, 'q': 1.5, 'p': 3, 'v': 8},
    'TOV': {'type': 'cost', 'k': 2.43, 'q': 1, 'p': 1.5, 'v': 2.5},
    'PF':  {'type': 'cost', 'k': 1, 'q': 1, 'p': 1.5, 'v': 2},
    '+/-': {'type': 'gain', 'k': 1.71, 'q': 3, 'p': 4, 'v': 8},
}
```

9. Czy istnieją jakieś dominacje w zbiorze wariantów? Jeśli tak, podaj je.

W zbiorze istnieją warianty zdominowane przez inne.

```python
for alt_1_id in df.index:
    for alt_2_id in df.index:
        if alt_1_id == alt_2_id:
            continue

        is_at_least_as_good_on_all = all([df.loc[alt_1_id, cr] <= df.loc[alt_2_id, cr] for cr, config in CRITERIA.items() if config['type'] == 'gain']) \
               and all([df.loc[alt_1_id, cr] >= df.loc[alt_2_id, cr] for cr, config in CRITERIA.items() if config['type'] == 'cost'])
        is_better_on_one = any([df.loc[alt_1_id, cr] < df.loc[alt_2_id, cr] for cr, config in CRITERIA.items() if config['type'] == 'gain']) \
                or any([df.loc[alt_1_id, cr] > df.loc[alt_2_id, cr] for cr, config in CRITERIA.items() if config['type'] == 'cost'])
        
        if is_at_least_as_good_on_all and is_better_on_one:
            print(f"{df.loc[alt_2_id, 'PLAYER'].ljust(23)} dominates {df.loc[alt_1_id, 'PLAYER']}")
```
